我们可以学习一些 JupyterLab 内置隐藏功能来简化工作，相信这些技巧对你有一定的帮助！
1. 窗口管理

    JupyterLab 具有内置的窗口管理器（无需第三方扩展）

    我们可以并排 Jupyter 笔记本(甚至垂直或水平组织三个笔记本)。通过拖拽的方式，可以自由的添加视图。



2. 并排

    在使用具有许多绘图或多行代码时，并排视图将会派上用场。

    并排可并排具有同一 Jupyter 笔记本的两个视图。这减少了不必要的滚动。

    要并排查看 Jupyter 笔记本，只需右键单击笔记本的标题并选择「Notebook 的新视图」
    
    还可以将输出的结果作为单独窗口查看，只需右键输出的结果选择「为输出创建新视图」

3. 拖拽 Cell

    鼠标在单元格前面显示成移动样式就可以任意拖动单元格

    每一个 Cell 都是可以拖拽的，不仅可以在单个文件内拖拽，而且可以在文件间进行拖拽，也支持一次拖拽多个单元。

4. 展开和收缩 Cell
    
    每一个 Cell 和 Output 左边都有一个蓝色的线，点击该蓝线，可以收拢或者展开，如果输出内容很多或者我们暂时不关心一些 cell 的内容时，就可以将其收拢起来。
    
    滚动输出，可以在 Cell / Output 上单击右键，然后选择 启用滚动输出 来固定内容展示高度，并启用滚动条。
    
    
5. Markdown 文件，可以点击右键菜单，选择预览进行查看。

6. 可以执行文本中的代码块
    
    例如：在 markdown 文档中，有一段 Python 代码，点击右键选择「为编辑器创建控制台」，点击代码片段任意位置后执行（Shift+Center）
    
    

7. 在 Jupyter 笔记本中安装包

    你可以直接从 Jupyter 中安装 Python 包。假设有一个缺少的 Python 包，只需在单元格中运行它即可：
```python
pip install numpy
```


8. 执行 Shell 命令

    Jupyter 能够执行 Shell 命令，只需要在 shell 命令前加一个感叹号!，Jupyter 会将其转换为 Bash。在任一命令前加感叹号!，它们就可以在 Python Jupyter Notebook 中运行。
    
9. 多行同时编辑（多光标支持）

    多行代码可以同时进行编辑，绝对的省时省力。按下`alt`键，利用鼠标左键进行选择
    
10. 隐藏部分输出信息

    在语句的最后添加`;`，可以去掉部分恼人的输出信息。
